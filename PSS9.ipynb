### No filter

```python
k = loadtxt("quadra.txt",usecols=0,skiprows=3)
T = loadtxt("quadra.txt",usecols=4,skiprows=3)
```

```python
plot(k,T)
```

```python
sup = []
j = []
L = len(k)
for i in range(5,L-5):
    if((T[i]>T[i-1]) & (T[i]>T[i-2]) & (T[i]>T[i-3]) & (T[i]>T[i-4]) & (T[i]>T[i-5]) & (T[i]>T[i+1]) & (T[i]>T[i+2]) & (T[i]>T[i+3]) & (T[i]>T[i+4]) & (T[i]>T[i+5]) & (k[i]<400) & (T[i]>20000)):
        sup.append(T[i])
        j.append(float(k[i]))    
```

```python
xlabel("$K$")
ylabel("|$\,$T$\,$|")
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plot(k,T)
plot(j,sup,'ro')

savefig("rumore_analisi.pdf")
```

### Analisi medie

### Non filtrato

```python
string = ["rumoreBuono2.txt","rumoreBuono3.txt","rumoreBuono4.txt","rumoreBuono5.txt"]
S = loadtxt("rumoreBuono.txt",usecols=4,skiprows=3)
S = S**2
```

```python
for i in range(0,4):
    T = loadtxt(string[i],usecols=4,skiprows=3)
    for j in range(0,len(T)):
        S[j] += T[j]**2
```

```python
S /= 5
S = sqrt(S)
```

```python
xlabel("$K$")
ylabel("S")
plot(S,'.')

savefig("studioRumore.pdf")
```

### Filtrato

```python
string = ["rumoreFiltro2.txt","rumoreFiltro3.txt","rumoreFiltro4.txt","rumoreFiltro5.txt"]
S = loadtxt("rumoreFiltro.txt",usecols=4,skiprows=3)
S = S**2
```

```python
for i in range(0,4):
    T = loadtxt(string[i],usecols=4,skiprows=3)
    for j in range(0,len(T)):
        S[j] += T[j]**2
```

```python
S /= 5
S = sqrt(S)
S = S[:-1]
```

```python
xlabel("$K$")
ylabel("S")
plot(S,'.')

savefig("studioRumoreFiltrato.pdf")
```
